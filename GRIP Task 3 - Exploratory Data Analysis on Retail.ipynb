## 3. Histogram Plots

```python
# finding nature of data through histogram plots
store.hist(bins=50 ,figsize=(20,15))
plt.show()
```

#### As these histograms show, the data is not normal; it is quite spread out.

```python
# comparing categories with sub-categories
plt.figure(figsize = (16, 8))
plt.bar('Sub-Category', 'Category', data=store)
plt.title('Category vs Sub-Category', size=25)
plt.xticks(rotation=90, size=15)
plt.yticks(size=15)
plt.show()
```

## 4. Profits in States

```python
# state counts
print('Total number of States are', store['State'].nunique())
print(store['State'].value_counts())
```

```python
# sorting the profits of states in decreasing order
states_profit = store.groupby(by='State').sum()['Profit'].sort_values(ascending=False).reset_index()
```

```python
# plotting top 10 states with highest profits
plt.figure(figsize = (15,10))
plt.title('States with Maximum Profits', size=30)
sns.barplot(x='State', y='Profit', data=states_profit.head(10)) 
plt.xticks(rotation=60, size=15)
plt.yticks(size=15)
plt.xlabel('States', size=20)
plt.ylabel('Profits', size=20)
plt.show()
```

```python
# plotting bottom 10 states with least profits
plt.figure(figsize = (15,10))
plt.title('States with Minimum Profits', size=30)
sns.barplot(x='State', y='Profit', data=states_profit.tail(10)) 
plt.xticks(rotation=60, size=15)
plt.yticks(size=15)
plt.xlabel('States', size=20)
plt.ylabel('Profits', size=20)
plt.show()
```

#### The most profitable state is California and the least profitable state is Texas.

## 5. Profits in Cities

```python
# sorting the profits of cities in decreasing order
cities_profit = store.groupby(by='City').sum()['Profit'].sort_values(ascending=False).reset_index()
```

```python
# plotting top 10 cities with highest profits
plt.figure(figsize = (15,10))
plt.title('Cities with Maximum Profits', size=30)
sns.barplot(x='City', y='Profit', data=cities_profit.head(10)) 
plt.xticks(rotation=60, size=15)
plt.yticks(size=15)
plt.xlabel('Cities', size=20)
plt.ylabel('Profits', size=20)
plt.show()
```

```python
# plotting bottom 10 cities with least profits
plt.figure(figsize = (15,10))
plt.title('Cities with Minimum Profits', size=30)
sns.barplot(x='City', y='Profit', data=cities_profit.tail(10)) 
plt.xticks(rotation=60, size=15)
plt.yticks(size=15)
plt.xlabel('Cities', size=20)
plt.ylabel('Profits', size=20)
plt.show()
```

#### The most profitable city is New York City and the least profitable city is Philadelphia.

## 6. Category Plots

```python
sns.countplot(x='Category', data=store)
plt.title('Category counts', size=20);
```

#### Office Supplies are in high demand. Technology category has the least supply.

```python
category_profit = store.groupby('Category').sum()['Profit'].sort_values(ascending = False)

plt.figure(figsize=(6,6))
plt.title('Profits in terms of Categories', size=20)
category_profit.plot(kind='bar')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

#### Technology category is the most profitable, despite having the least supply. Furniture category gains much less profit as compared to other two sectors.

```python
sales = store.groupby('Category').sum()['Sales'].nlargest(n=10) # top 10 states with highest sales
profits = store.groupby('Category').sum()['Profit'].nlargest(n=10) # top 10 states with highest profits

plt.style.use('seaborn')
sales.plot(kind='bar', fontsize=15, color='c')
profits.plot(kind='bar', fontsize=15, color='m')
plt.xlabel('Categories', fontsize=15)
plt.ylabel('Total Sales and Profits', fontsize=15)
plt.title("Sales vs Profits in terms of Categories", size=20)
plt.show()
```

#### Technology and Office Supplies have high sales and profit. 

## 7. Sub-Category Plots

```python
plt.figure(figsize = (16,8))
sns.countplot(x='Sub-Category', data=store)
plt.title('Sub-Category counts', size=20);
```

#### Binders and Paper have very heavy demand. Machines and Copiers have the least demand.

```python
sub_profit = store.groupby('Sub-Category').sum()['Profit'].sort_values(ascending = False)

plt.figure(figsize=(16,6))
plt.title('Profits in terms of Sub-Categories', size=20)
sub_profit.plot(kind='bar')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

#### Copiers have highest profit, while Tables have lowest profit. 

```python
min_profit = store.groupby('Sub-Category').min()['Profit'].sort_values()

plt.figure(figsize = (8,6))
min_profit.plot()
plt.title('Sub-Categories with Highest Losses', size=20)
plt.xticks(rotation=60, fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

#### Machines have incurred more losses than other sub-categories.

## 8. Shipping Mode Plots 

```python
sns.countplot(x='Ship Mode', data=store)
plt.title('Shipping Mode counts', size=20);
```

#### Standard Class is the most preferred shipping mode. 

```python
ship_profit = store.groupby('Ship Mode').sum()['Profit'].sort_values(ascending = False)

plt.figure(figsize=(16,6))
plt.title('Profits in terms of Shipping Modes', size=20)
ship_profit.plot(kind='bar')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

#### Standard Class shipping mode shows very high profit as compared to others.

```python
ship_profit_avg = store.groupby('Ship Mode').mean()['Profit'].sort_values(ascending = False)

plt.figure(figsize=(16,6))
plt.title('Average Profits in terms of Shipping Modes', size=20)
ship_profit_avg.plot(kind='bar')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
ship_discount = store.groupby('Ship Mode').sum()['Discount'].sort_values(ascending = False).reset_index()

plt.figure(figsize=(10,6))
plt.title('Discounts in terms of Shipping Modes', size=20)
sns.barplot(x='Ship Mode', y='Discount', data=ship_discount)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

#### Standard Class shipping mode provides maximum discount and gains maximum profit. But the average profit is minimum.

## 9. Segment Plots

```python
sns.countplot(x='Segment', data=store)
plt.title('Segment counts', size=20);
```

#### Consumer orders are the most and Home Office orders are the least.

```python
segment_profit = store.groupby('Segment').mean()['Profit'].sort_values(ascending = False).reset_index()

plt.figure(figsize=(8,5))
plt.title('Profits according to Segments', size=20)
sns.barplot(x='Segment', y='Profit', data=segment_profit)
plt.xticks(rotation=60 , fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
sales = store.groupby('Segment').mean()['Sales'].sort_values(ascending = False)
profits = store.groupby('Segment').mean()['Profit'].sort_values(ascending = False)

plt.style.use('seaborn')
sales.plot(kind='bar', fontsize=15, color='r')
profits.plot(kind='bar', fontsize=15, color='g')
plt.xlabel('Segments',fontsize=15)
plt.ylabel('Total Sales and Profits', fontsize=15)
plt.title("Sales vs Profits in terms of Segments", size=20)
plt.show()
```

#### Home Office segment has more sales and profits, inspite of having less orders. Consumer segment has the least sales and profit.

## 10. Region Plots

```python
sns.countplot(x='Region', data=store)
plt.title('Region counts', size=20);
```

#### Orders are delivered more in the West and East zones. 

```python
regions_profit = store.groupby('Region').sum()['Profit'].sort_values(ascending = False).reset_index()

plt.figure(figsize=(10,5))
plt.title('Profits according to Regions', size=20)
sns.barplot(x='Region', y='Profit', data=regions_profit)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
sales = store.groupby('Region').sum()['Sales'].sort_values(ascending = False)
profits = store.groupby('Region').sum()['Profit'].sort_values(ascending = False)

plt.style.use('seaborn')
sales.plot(kind='bar', fontsize=15, color='b')
profits.plot(kind='bar', fontsize=15, color='y')
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Total Sales and Profits', fontsize=15)
plt.title("Sales vs Profits in terms of Regions", size=20)
plt.show()
```

#### West and East zones have more sales and profit.

## 11. Sales, Profits and Discounts

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Discount', y='Profit', data=store, label='Discount')
plt.title("Profit vs Discount", size=20)
plt.legend()
plt.show()
```

#### Profit and Discount have a negative correlation.

```python
sns.lineplot(x='Sales', y='Profit', data=store)
plt.title("Profit vs Sales", size=20)
plt.show()
```

#### Profit and Sales have a moderately strong correlation.

```python
sales = store.groupby('State').sum()['Sales'].nlargest(n=10) # top 10 states with highest sales
profits = store.groupby('State').sum()['Profit'].nlargest(n=10) # top 10 states with highest profits

sales.plot(kind='bar', fontsize=15, color='purple')
profits.plot(kind='bar', fontsize=15 , color='pink')
plt.xlabel('States', fontsize=15)
plt.ylabel('Total sales and profits', fontsize=15)
plt.title("Sales vs Profit in States", size=20)
plt.show()
```

#### From the plot, we find that California and New York states have the highest sales and profits.

```python
plt.style.use('seaborn')
store.plot(kind="scatter", x="Sales", y="Profit", c="Discount", s=20, fontsize=16, colormap="Spectral")
plt.ylabel('Total Profits', fontsize=15)
plt.title("Sales vs Profits on Discounts", fontsize=20)
plt.show()
```

#### Sales increase diminishingly with increase in discounts, but profit decreases drastically.

```python
plt.figure(figsize = (7,6))
sns.kdeplot(store['Sales'], color='b', label='Sales', shade=True, bw_method=20)
sns.kdeplot(store['Profit'], color='m', label='Profit', shade=True, bw_method=20)
plt.title('KDE Plot of Sales and Profit', size=20)
plt.xlabel('Sales/Profit', size=15)
plt.ylabel('Density', size=15)
plt.legend()
plt.show()
```

#### Profit is more than Sales for a given value.
